```python
from tika import parser
import os
import sys
import io
import pickle
import textract

UU_Title= []
UU_Content = []

def extract_text_from_pdfs_recursively(dir):
    for foldername, subfolders, files in os.walk(dir,"rb"):
        for file in files:
            tmp_pdf = foldername+"/"+file
            path_to_pdf = os.path.join(foldername, file)
            [stem, ext] = os.path.splitext(path_to_pdf)
            if ext == '.pdf':
                print("Processing " + tmp_pdf)
                pdf_contents = parser.from_file(path_to_pdf)
                UU_Title.append(pdf_contents['metadata']['resourceName'])
                UU_Content.append(pdf_contents['content'])
                 #print (pdf_contents)
                

# extracting pdf file
extract_text_from_pdfs_recursively("D:/Portofolio/Indonesian-Controversial-Laws-Wordcloud/LawsDocument")
```

```python
# to replace title with format that we want
UU_Title = [d.replace("b'","") for d in UU_Title]
UU_Title = [d.replace(".pdf'","") for d in UU_Title]
```

```python
# combining UU_title with UU_content
UU_Title_Content = {UU_Title[i]: [UU_Content[i]] for i in range(len(UU_Content))}
UU_Title_Content.keys()
```

```python
# Collect and then removing none value
# optional
law_unextract = []
for key, value in UU_Title_Content.items():
    if UU_Title_Content[key] == [None] :
        #print(key)
        law_unextract.append(key)
#len(law_unextract)

# saving unextracted document lists
with open('D:/Portofolio/Indonesian-Controversial-Laws-Wordcloud/pickle/law_unextract.pkl', 'wb') as un: 
    pickle.dump(law_unextract, un)
    
# saving extracted title documents
extracted_title = [t for t in UU_Title if t not in law_unextract]
with open('D:/Portofolio/Indonesian-Controversial-Laws-Wordcloud/pickle/extracted_title.pkl', 'wb') as t:
    pickle.dump(extracted_title, t)
len(extracted_title)
```

```python
# Merubah dictionary menjadi dataframe untuk preprosesing
import pandas as pd
pd.set_option('max_colwidth',150)
# uu_df adalah untuk dataframe undang-undang yang dipisahkan berdasarkan kategori yang ada
extracted_law_df = pd.DataFrame.from_dict(UU_Title_Content).transpose()
extracted_law_df.columns = ['Top_6']
extracted_law_df = extracted_law_df.sort_index()
extracted_law_df
```

```python
# firt preprocessing
"""text lowercasing, disappearing number and punctuation."""
import re
import string

def preproses_tahap1(text):
    text = text.lower()
    text = re.sub('www.bphn.go.id','',text)
    text = re.sub('www.hukumonline.com','',text)
    #untuk menghilangkan teks yang berada di dalam []
    text = re.sub('\[.*?\]', '', text)
    #untuk menghilangkan karakter non ascii
    text = text.encode('ascii', 'replace').decode('ascii')
    #Untuk menghilangkan punktuasi (tanda baca) yang berada di dalam teks
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    #untuk menghilangkan kata yang terdapat angka di dalammnya
    text = re.sub('\w*\d\w*', '', text)
    return text

preptahap1 = lambda x: preproses_tahap1(x)
law_prep = pd.DataFrame(extracted_law_df.Top_6.apply(preptahap1))
law_prep.columns = ['Top_6']
law_prep
```

```python
# second preprocessing 
"""completing first preprocessing"""
from collections import Counter
def preproses_tahap2(text):
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\t', ' ', text)
    text = text.strip()
    text = re.sub('microsoft word', '', text)
    text = re.sub('�', '', text)
    text = re.sub('\s+',' ',text)
    #untuk menghilangkan single char
    text = re.sub(r"\b[a-zA-Z]\b", "", text)
    text = re.sub('copyright','',text)
    text = re.sub('menirr', '', text)
    text = re.sub('nomor', '', text)
    text = re.sub('repuel', '', text)
    text = re.sub('indonesta', 'indonesia', text)
    text = re.sub('teks format asli kembali tekan backspace','', text)
    text = re.sub('republtk', 'republik', text)
    return text
    
preptahap2 = lambda x: preproses_tahap2(x)
law_prep = pd.DataFrame(law_prep.Top_6.apply(preptahap2))
law_prep = pd.DataFrame(law_prep.Top_6.apply(lambda x: ' '.join([word for word in x.split() if len(word)>3])))
law_prep = pd.DataFrame(law_prep.Top_6.apply(lambda x: ' '.join([word for word in x.split() if x.count(word)<5])))

# Menyimpan hasil preprosing tahap 2 ke dalam bentuk pickle
law_prep.to_pickle('D:/Portofolio/Indonesian-Controversial-Laws-Wordcloud/pickle/law_prep.pkl')
law_prep
```

```python
# Untuk menghilangkan stopwords menggunakan nltk stopwords
from nltk.corpus import stopwords as sw
s_word = sw.words('indonesian','english')
#extend dengan nama bulan
s_word.extend(['januari','februari','maret','april','mei','juni','juli','agustus','september','oktober','november','desember'])

law_stoprem = pd.DataFrame(law_prep.Top_6.apply(lambda x: ' '.join([word for word in x.split() if word not in (s_word)])))
law_stoprem.columns = ["Top_6"]
law_stoprem
```
